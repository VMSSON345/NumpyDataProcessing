```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Đọc dữ liệu từ file activity_period.csv
df_activity = pd.read_csv('DATA_PYTHON/data_processed/activity_periods.csv')

# Đọc dữ liệu từ file contributor.csv
df_contributor = pd.read_csv('DATA_PYTHON/data_processed/contributors.csv')

# Kết hợp dữ liệu theo 'author' (hoặc 'login' tùy theo tên cột)
df_combined = pd.merge(df_activity, df_contributor, left_on='author', right_on='login', how='inner')

# Kiểm tra dữ liệu kết hợp
print(df_combined.head())

# 1. Biểu đồ phân tán 3D giữa 'activity_duration', 'commits', và 'percent_contribution'
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Vẽ các điểm phân tán
scatter = ax.scatter(df_combined['activity_duration'], df_combined['commits'], df_combined['percent_contribution'], c='orange', marker='o')

# Gán nhãn cho các trục
ax.set_xlabel('Thời gian hoạt động (giờ)')
ax.set_ylabel('Số lượng commit')
ax.set_zlabel('Percent Contribution')

# Tiêu đề
plt.title('Biểu đồ phân tán 3D giữa thời gian hoạt động, số lượng commit và percent contribution')

# Hiển thị biểu đồ
plt.show()

# 2. Biểu đồ phân tán giữa 'activity_duration' và 'commits'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['activity_duration'], y=df_combined['commits'], color='green', s=100)
plt.title("Mối quan hệ giữa Thời gian hoạt động và Số lượng commit")
plt.xlabel("Thời gian hoạt động (giờ)")
plt.ylabel("Số lượng commit")
plt.show()

# 3. Biểu đồ phân tán giữa 'percent_contribution' và 'commits'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['percent_contribution'], y=df_combined['commits'], color='blue', s=100)
plt.title("Mối quan hệ giữa Percent Contribution và Số lượng commit")
plt.xlabel("Percent Contribution")
plt.ylabel("Số lượng commit")
plt.show()

# 4. Biểu đồ phân tán giữa 'percent_contribution' và 'activity_duration'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['percent_contribution'], y=df_combined['activity_duration'], color='red', s=100)
plt.title("Mối quan hệ giữa Percent Contribution và Thời gian hoạt động")
plt.xlabel("Percent Contribution")
plt.ylabel("Thời gian hoạt động (giờ)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Violin plot cho activity_duration, commits và percent_contribution
plt.figure(figsize=(12, 6))

# Vẽ violin plot cho activity_duration
plt.subplot(1, 3, 1)
sns.violinplot(data=df_combined, x='activity_duration', color='skyblue')
plt.title('Phân bố Thời gian Hoạt động')

# Vẽ violin plot cho commits
plt.subplot(1, 3, 2)
sns.violinplot(data=df_combined, x='commits', color='lightgreen')
plt.title('Phân bố Số lượng Commit')

# Vẽ violin plot cho percent_contribution
plt.subplot(1, 3, 3)
sns.violinplot(data=df_combined, x='percent_contribution', color='lightcoral')
plt.title('Phân bố Percent Contribution')

plt.tight_layout()
plt.show()
```
