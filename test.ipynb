```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# c d liu t file activity_period.csv
df_activity = pd.read_csv('DATA_PYTHON/data_processed/activity_periods.csv')

# c d liu t file contributor.csv
df_contributor = pd.read_csv('DATA_PYTHON/data_processed/contributors.csv')

# Kt hp d liu theo 'author' (hoc 'login' ty theo tn ct)
df_combined = pd.merge(df_activity, df_contributor, left_on='author', right_on='login', how='inner')

# Kim tra d liu kt hp
print(df_combined.head())

# 1. Biu  phn tn 3D gia 'activity_duration', 'commits', v 'percent_contribution'
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# V cc im phn tn
scatter = ax.scatter(df_combined['activity_duration'], df_combined['commits'], df_combined['percent_contribution'], c='orange', marker='o')

# Gn nhn cho cc trc
ax.set_xlabel('Thi gian hot ng (gi)')
ax.set_ylabel('S lng commit')
ax.set_zlabel('Percent Contribution')

# Tiu 
plt.title('Biu  phn tn 3D gia thi gian hot ng, s lng commit v percent contribution')

# Hin th biu 
plt.show()

# 2. Biu  phn tn gia 'activity_duration' v 'commits'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['activity_duration'], y=df_combined['commits'], color='green', s=100)
plt.title("Mi quan h gia Thi gian hot ng v S lng commit")
plt.xlabel("Thi gian hot ng (gi)")
plt.ylabel("S lng commit")
plt.show()

# 3. Biu  phn tn gia 'percent_contribution' v 'commits'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['percent_contribution'], y=df_combined['commits'], color='blue', s=100)
plt.title("Mi quan h gia Percent Contribution v S lng commit")
plt.xlabel("Percent Contribution")
plt.ylabel("S lng commit")
plt.show()

# 4. Biu  phn tn gia 'percent_contribution' v 'activity_duration'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_combined['percent_contribution'], y=df_combined['activity_duration'], color='red', s=100)
plt.title("Mi quan h gia Percent Contribution v Thi gian hot ng")
plt.xlabel("Percent Contribution")
plt.ylabel("Thi gian hot ng (gi)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Violin plot cho activity_duration, commits v percent_contribution
plt.figure(figsize=(12, 6))

# V violin plot cho activity_duration
plt.subplot(1, 3, 1)
sns.violinplot(data=df_combined, x='activity_duration', color='skyblue')
plt.title('Phn b Thi gian Hot ng')

# V violin plot cho commits
plt.subplot(1, 3, 2)
sns.violinplot(data=df_combined, x='commits', color='lightgreen')
plt.title('Phn b S lng Commit')

# V violin plot cho percent_contribution
plt.subplot(1, 3, 3)
sns.violinplot(data=df_combined, x='percent_contribution', color='lightcoral')
plt.title('Phn b Percent Contribution')

plt.tight_layout()
plt.show()
```
